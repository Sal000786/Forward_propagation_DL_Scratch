# importing numpy for numerical operation

```python
import numpy as np
```

# sigmoid function to use as activation for NN(Neural network)

```python
def sigmoid(x):
    result=1/(1+np.exp(-x))
    return result
```

```python
sigmoid(4)
```

# first layer of the network started

# Input layer with 4  inputs

```python
inputs=np.random.random(4)
inputs
```

```python
inputs=inputs.reshape(4,1)
inputs.shape
```

# 1st hidden layer with 3 nodes

```python
hidden_1=3
```

# weights of the first layer (4,3) 

```python
weights_1=np.random.random((inputs.shape[0],hidden_1))
```

```python
weights_1
```

```python
weights_1.shape
```

# transpose the weights matrix for dot product compatibility

```python
weights_1_trans=np.transpose(weights_1)
weights_1_trans.shape
```

# this is similar to w.x(i.e weights of layer 1 dot product with inputs)

```python
mult_1=np.dot(weights_1_trans,inputs)
mult_1
```

```python
mult_1.shape
```

# biases for the layer 1

```python
bias_1=np.ones(hidden_1)
bias_1
```

```python
bias_1=bias_1.reshape(3,1)
bias_1.shape
```

# This is similar to wx+b that will be passed to activation function

```python
wx_b_1=mult_1+bias_1
wx_b_1
```

# ouptu of the first layer

```python
output_1=sigmoid(wx_b_1)
output_1
```

```python
output_1.shape
```

# Layer 1 completed

# 2nd layer of the network started from here

# Repeat the same process with different size of the weights and inputs, biases as explained in the NN architecture

```python
hidden_2=2
```

```python
weights_2=np.random.random((output_1.shape[0],hidden_2))
weights_2.shape
```

```python
weights_2
```

```python
bias_2=np.ones(hidden_2)
bias_2
```

```python
bias_2=bias_2.reshape(2,1)
bias_2.shape
```

```python
mult_2=np.dot(np.transpose(weights_2),output_1)
mult_2
```

```python
wx_b_2=mult_2+bias_2
wx_b_2
```

```python
output_2=sigmoid(wx_b_2)
output_2
```

# Layer 2 completed

# Layer 3 started

```python
hidden_3=1
```

```python
weights_3=np.random.random((output_2.shape[0],hidden_3))
```

```python
weights_3
```

```python
bias_3=np.ones(hidden_3)
bias_3
```

```python
mult_3=np.dot(np.transpose(weights_3),output_2)
mult_3
```

```python
mult_3.shape
```

```python
wx_b_3=mult_3+bias_3
wx_b_3
```

```python
output_3=sigmoid(wx_b_3)
```

```python
output_3 # this is our final output
```

# finally our neural network is ready with 2 hidden layer 1 input layer and 1 ouptut layer from scratch
